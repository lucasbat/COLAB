```python
# ============================================================
# Comparação entre MLP baseline e CNNs pré-treinadas no Fashion-MNIST
# Modelos: MLP, MobileNetV2, EfficientNetB0 e DenseNet121
# Variações: dropout, otimizador e learning rate
# Google Colab
# ============================================================

import os
import gc
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

# ============================================================
# 1. Reprodutibilidade
# ============================================================

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

print("TensorFlow:", tf.__version__)

# ============================================================
# 2. Configurações gerais
# ============================================================

EPOCHS_MLP = 12
EPOCHS_TRANSFER = 10
BATCH_SIZE = 128
IMG_SIZE = 96

# Se quiser acelerar testes, mude para True
USE_SUBSET = False
SUBSET_TRAIN = 15000
SUBSET_VAL = 3000
SUBSET_TEST = 3000

# ============================================================
# 3. Carregamento do dataset
# ============================================================

(x_train_full, y_train_full), (x_test, y_test) = fashion_mnist.load_data()

class_names = [
    "Camiseta/top", "Calça", "Pulôver", "Vestido", "Casaco",
    "Sandália", "Camisa", "Tênis", "Bolsa", "Bota"
]

print("Treino original:", x_train_full.shape)
print("Teste:", x_test.shape)

# ============================================================
# 4. Normalização e divisão treino/validação
# ============================================================

x_train_full = x_train_full.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

x_train = x_train_full[:50000]
y_train = y_train_full[:50000]

x_val = x_train_full[50000:]
y_val = y_train_full[50000:]

if USE_SUBSET:
    x_train = x_train[:SUBSET_TRAIN]
    y_train = y_train[:SUBSET_TRAIN]

    x_val = x_val[:SUBSET_VAL]
    y_val = y_val[:SUBSET_VAL]

    x_test = x_test[:SUBSET_TEST]
    y_test = y_test[:SUBSET_TEST]

print("Treino usado:", x_train.shape)
print("Validação usada:", x_val.shape)
print("Teste usado:", x_test.shape)

# Para MLP
x_train_mlp = x_train.reshape(-1, 28 * 28)
x_val_mlp = x_val.reshape(-1, 28 * 28)
x_test_mlp = x_test.reshape(-1, 28 * 28)

# Para modelos convolucionais
x_train_cnn = x_train.reshape(-1, 28, 28, 1)
x_val_cnn = x_val.reshape(-1, 28, 28, 1)
x_test_cnn = x_test.reshape(-1, 28, 28, 1)

# Labels one-hot
y_train_cat = to_categorical(y_train, 10)
y_val_cat = to_categorical(y_val, 10)
y_test_cat = to_categorical(y_test, 10)

# ============================================================
# 5. Visualização de exemplos
# ============================================================

plt.figure(figsize=(10, 5))
for i in range(12):
    plt.subplot(3, 4, i + 1)
    plt.imshow(x_train[i], cmap="gray")
    plt.title(class_names[y_train[i]])
    plt.axis("off")

plt.suptitle("Exemplos do dataset Fashion-MNIST")
plt.tight_layout()
plt.savefig("figura_exemplos_dataset.png", dpi=300)
plt.show()

# ============================================================
# 6. Funções auxiliares
# ============================================================

def criar_callbacks():
    return [
        EarlyStopping(
            monitor="val_loss",
            patience=4,
            restore_best_weights=True
        ),
        ReduceLROnPlateau(
            monitor="val_loss",
            factor=0.5,
            patience=2,
            min_lr=1e-6
        )
    ]


def obter_otimizador(nome, learning_rate):
    nome = nome.lower()

    if nome == "adam":
        return tf.keras.optimizers.Adam(learning_rate=learning_rate)

    if nome == "rmsprop":
        return tf.keras.optimizers.RMSprop(learning_rate=learning_rate)

    if nome == "sgd":
        return tf.keras.optimizers.SGD(
            learning_rate=learning_rate,
            momentum=0.9
        )

    raise ValueError(f"Otimizador não reconhecido: {nome}")


def calcular_metricas(nome, arquitetura, dropout, optimizer, learning_rate, y_true, y_pred):
    return {
        "Modelo": nome,
        "Arquitetura": arquitetura,
        "Dropout": dropout,
        "Otimizador": optimizer,
        "Learning rate": learning_rate,
        "Acurácia": accuracy_score(y_true, y_pred),
        "Precisão macro": precision_score(y_true, y_pred, average="macro", zero_division=0),
        "Revocação macro": recall_score(y_true, y_pred, average="macro", zero_division=0),
        "F1-score macro": f1_score(y_true, y_pred, average="macro", zero_division=0)
    }


def plot_history(history, titulo, arquivo):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history["accuracy"], label="Treino")
    plt.plot(history.history["val_accuracy"], label="Validação")
    plt.title(f"Acurácia - {titulo}")
    plt.xlabel("Época")
    plt.ylabel("Acurácia")
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.subplot(1, 2, 2)
    plt.plot(history.history["loss"], label="Treino")
    plt.plot(history.history["val_loss"], label="Validação")
    plt.title(f"Perda - {titulo}")
    plt.xlabel("Época")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(arquivo, dpi=300)
    plt.show()


def plot_confusion(y_true, y_pred, titulo, arquivo):
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(9, 7))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=class_names,
        yticklabels=class_names
    )
    plt.title(titulo)
    plt.xlabel("Classe prevista")
    plt.ylabel("Classe real")
    plt.xticks(rotation=45, ha="right")
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.savefig(arquivo, dpi=300)
    plt.show()


def nome_arquivo_seguro(texto):
    return (
        texto.lower()
        .replace(" ", "_")
        .replace("+", "mais")
        .replace("/", "_")
        .replace("-", "_")
        .replace(".", "_")
    )

# ============================================================
# 7. Modelo baseline - MLP
# ============================================================

mlp_model = models.Sequential([
    layers.Input(shape=(784,)),

    layers.Dense(
        256,
        activation="relu",
        kernel_regularizer=l2(1e-4)
    ),
    layers.Dropout(0.4),

    layers.Dense(
        128,
        activation="relu",
        kernel_regularizer=l2(1e-4)
    ),
    layers.Dropout(0.4),

    layers.Dense(10, activation="softmax")
])

mlp_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

print("\nResumo do modelo MLP baseline:")
mlp_model.summary()

print("\nTreinando MLP baseline...")

history_mlp = mlp_model.fit(
    x_train_mlp,
    y_train_cat,
    validation_data=(x_val_mlp, y_val_cat),
    epochs=EPOCHS_MLP,
    batch_size=BATCH_SIZE,
    callbacks=criar_callbacks(),
    verbose=1
)

mlp_probs = mlp_model.predict(x_test_mlp)
mlp_pred = np.argmax(mlp_probs, axis=1)

resultados = []
predicoes = {}
historicos = {}

resultados.append(
    calcular_metricas(
        nome="MLP baseline",
        arquitetura="MLP",
        dropout=0.4,
        optimizer="Adam",
        learning_rate=0.001,
        y_true=y_test,
        y_pred=mlp_pred
    )
)

predicoes["MLP baseline"] = mlp_pred
historicos["MLP baseline"] = history_mlp

print("\nRelatório MLP baseline:")
print(classification_report(y_test, mlp_pred, target_names=class_names, zero_division=0))

pd.DataFrame(
    classification_report(
        y_test,
        mlp_pred,
        target_names=class_names,
        output_dict=True,
        zero_division=0
    )
).transpose().to_csv("relatorio_por_classe_mlp_baseline.csv")

plot_history(
    history_mlp,
    "MLP baseline",
    "curvas_mlp_baseline.png"
)

plot_confusion(
    y_test,
    mlp_pred,
    "Matriz de Confusão - MLP baseline",
    "matriz_confusao_mlp_baseline.png"
)

# ============================================================
# 8. Data augmentation
# ============================================================

data_augmentation = tf.keras.Sequential([
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomTranslation(0.1, 0.1)
], name="data_augmentation")

# ============================================================
# 9. Construção de modelos CNN pré-treinados
# ============================================================

def criar_modelo_transfer_learning(
    arquitetura,
    dropout_rate,
    optimizer_name,
    learning_rate
):
    inputs = layers.Input(shape=(28, 28, 1))

    x = data_augmentation(inputs)

    x = layers.Resizing(IMG_SIZE, IMG_SIZE)(x)

    x = layers.Lambda(
        lambda img: tf.image.grayscale_to_rgb(img),
        name="grayscale_to_rgb"
    )(x)

    if arquitetura == "MobileNetV2":
        from tensorflow.keras.applications import MobileNetV2
        from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

        preprocess = layers.Lambda(
            lambda img: preprocess_input(img * 255.0),
            name="mobilenetv2_preprocessing"
        )

        base_model = MobileNetV2(
            include_top=False,
            weights="imagenet",
            input_shape=(IMG_SIZE, IMG_SIZE, 3)
        )

    elif arquitetura == "EfficientNetB0":
        from tensorflow.keras.applications import EfficientNetB0
        from tensorflow.keras.applications.efficientnet import preprocess_input

        preprocess = layers.Lambda(
            lambda img: preprocess_input(img * 255.0),
            name="efficientnetb0_preprocessing"
        )

        base_model = EfficientNetB0(
            include_top=False,
            weights="imagenet",
            input_shape=(IMG_SIZE, IMG_SIZE, 3)
        )

    elif arquitetura == "DenseNet121":
        from tensorflow.keras.applications import DenseNet121
        from tensorflow.keras.applications.densenet import preprocess_input

        preprocess = layers.Lambda(
            lambda img: preprocess_input(img * 255.0),
            name="densenet121_preprocessing"
        )

        base_model = DenseNet121(
            include_top=False,
            weights="imagenet",
            input_shape=(IMG_SIZE, IMG_SIZE, 3)
        )

    else:
        raise ValueError(f"Arquitetura não reconhecida: {arquitetura}")

    base_model.trainable = False

    x = preprocess(x)

    x = base_model(x, training=False)

    x = layers.GlobalAveragePooling2D()(x)

    x = layers.BatchNormalization()(x)

    x = layers.Dropout(dropout_rate)(x)

    x = layers.Dense(
        128,
        activation="relu",
        kernel_regularizer=l2(1e-4)
    )(x)

    x = layers.Dropout(dropout_rate / 2)(x)

    outputs = layers.Dense(10, activation="softmax")(x)

    model = models.Model(inputs, outputs)

    model.compile(
        optimizer=obter_otimizador(optimizer_name, learning_rate),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )

    return model

# ============================================================
# 10. Experimentos com CNNs pré-treinadas
# ============================================================

experimentos = [
    {
        "nome": "MobileNetV2 - Dropout 0.30 - Adam - LR 1e-3",
        "arquitetura": "MobileNetV2",
        "dropout": 0.30,
        "optimizer": "Adam",
        "learning_rate": 1e-3
    },
    {
        "nome": "MobileNetV2 - Dropout 0.50 - Adam - LR 5e-4",
        "arquitetura": "MobileNetV2",
        "dropout": 0.50,
        "optimizer": "Adam",
        "learning_rate": 5e-4
    },
    {
        "nome": "EfficientNetB0 - Dropout 0.30 - Adam - LR 1e-3",
        "arquitetura": "EfficientNetB0",
        "dropout": 0.30,
        "optimizer": "Adam",
        "learning_rate": 1e-3
    },
    {
        "nome": "EfficientNetB0 - Dropout 0.50 - RMSprop - LR 1e-4",
        "arquitetura": "EfficientNetB0",
        "dropout": 0.50,
        "optimizer": "RMSprop",
        "learning_rate": 1e-4
    },
    {
        "nome": "DenseNet121 - Dropout 0.30 - Adam - LR 5e-4",
        "arquitetura": "DenseNet121",
        "dropout": 0.30,
        "optimizer": "Adam",
        "learning_rate": 5e-4
    },
    {
        "nome": "DenseNet121 - Dropout 0.50 - SGD - LR 1e-3",
        "arquitetura": "DenseNet121",
        "dropout": 0.50,
        "optimizer": "SGD",
        "learning_rate": 1e-3
    }
]

for exp in experimentos:
    print("\n" + "=" * 80)
    print("Treinando:", exp["nome"])
    print("=" * 80)

    tf.keras.backend.clear_session()
    gc.collect()

    model = criar_modelo_transfer_learning(
        arquitetura=exp["arquitetura"],
        dropout_rate=exp["dropout"],
        optimizer_name=exp["optimizer"],
        learning_rate=exp["learning_rate"]
    )

    print("\nResumo do modelo:")
    model.summary()

    history = model.fit(
        x_train_cnn,
        y_train_cat,
        validation_data=(x_val_cnn, y_val_cat),
        epochs=EPOCHS_TRANSFER,
        batch_size=BATCH_SIZE,
        callbacks=criar_callbacks(),
        verbose=1
    )

    probs = model.predict(x_test_cnn)
    y_pred = np.argmax(probs, axis=1)

    resultados.append(
        calcular_metricas(
            nome=exp["nome"],
            arquitetura=exp["arquitetura"],
            dropout=exp["dropout"],
            optimizer=exp["optimizer"],
            learning_rate=exp["learning_rate"],
            y_true=y_test,
            y_pred=y_pred
        )
    )

    predicoes[exp["nome"]] = y_pred
    historicos[exp["nome"]] = history

    print(f"\nRelatório por classe - {exp['nome']}:")
    print(classification_report(y_test, y_pred, target_names=class_names, zero_division=0))

    nome_seguro = nome_arquivo_seguro(exp["nome"])

    pd.DataFrame(
        classification_report(
            y_test,
            y_pred,
            target_names=class_names,
            output_dict=True,
            zero_division=0
        )
    ).transpose().to_csv(f"relatorio_por_classe_{nome_seguro}.csv")

    plot_history(
        history,
        exp["nome"],
        f"curvas_{nome_seguro}.png"
    )

    plot_confusion(
        y_test,
        y_pred,
        f"Matriz de Confusão - {exp['nome']}",
        f"matriz_confusao_{nome_seguro}.png"
    )

    del model
    gc.collect()

# ============================================================
# 11. Tabela final de resultados
# ============================================================

df_resultados = pd.DataFrame(resultados)

df_resultados = df_resultados.sort_values(
    by="F1-score macro",
    ascending=False
).reset_index(drop=True)

df_resultados_formatado = df_resultados.copy()

for col in ["Dropout", "Learning rate", "Acurácia", "Precisão macro", "Revocação macro", "F1-score macro"]:
    df_resultados_formatado[col] = df_resultados_formatado[col].map(lambda x: f"{x:.4f}")

print("\nTabela comparativa final:")
display(df_resultados_formatado)

df_resultados.to_csv("tabela_metricas_modelos_transfer_learning.csv", index=False)

# ============================================================
# 12. Gráfico comparativo das métricas
# ============================================================

plt.figure(figsize=(14, 6))

df_plot = df_resultados.copy()
df_plot["Modelo curto"] = df_plot["Modelo"].str.replace(" - ", "\n", regex=False)

sns.barplot(
    data=df_plot,
    x="Modelo curto",
    y="F1-score macro",
    hue="Arquitetura"
)

plt.title("Comparação dos modelos por F1-score macro")
plt.xlabel("Modelo")
plt.ylabel("F1-score macro")
plt.ylim(0, 1)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Arquitetura")
plt.tight_layout()
plt.savefig("comparacao_f1_score_modelos.png", dpi=300)
plt.show()

plt.figure(figsize=(14, 6))

sns.barplot(
    data=df_plot,
    x="Modelo curto",
    y="Acurácia",
    hue="Arquitetura"
)

plt.title("Comparação dos modelos por acurácia")
plt.xlabel("Modelo")
plt.ylabel("Acurácia")
plt.ylim(0, 1)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Arquitetura")
plt.tight_layout()
plt.savefig("comparacao_acuracia_modelos.png", dpi=300)
plt.show()

# ============================================================
# 13. Melhor modelo e análise de erros
# ============================================================

melhor_modelo = df_resultados.iloc[0]["Modelo"]
melhor_pred = predicoes[melhor_modelo]

print("\nMelhor modelo encontrado:")
print(melhor_modelo)

plot_confusion(
    y_test,
    melhor_pred,
    f"Matriz de Confusão - Melhor modelo: {melhor_modelo}",
    "matriz_confusao_melhor_modelo.png"
)

erros = np.where(melhor_pred != y_test)[0]

plt.figure(figsize=(12, 8))

for i, idx in enumerate(erros[:12]):
    plt.subplot(3, 4, i + 1)
    plt.imshow(x_test[idx], cmap="gray")
    plt.title(
        f"Real: {class_names[y_test[idx]]}\nPred: {class_names[melhor_pred[idx]]}",
        fontsize=9
    )
    plt.axis("off")

plt.suptitle(f"Exemplos de erros - {melhor_modelo}")
plt.tight_layout()
plt.savefig("exemplos_erros_melhor_modelo.png", dpi=300)
plt.show()

# ============================================================
# 14. Salvamento dos históricos de treinamento
# ============================================================

for nome_modelo, history in historicos.items():
    nome_seguro = nome_arquivo_seguro(nome_modelo)
    pd.DataFrame(history.history).to_csv(
        f"historico_treinamento_{nome_seguro}.csv",
        index=False
    )

# ============================================================
# 15. Resumo final
# ============================================================

print("\nArquivos principais gerados:")
print("- tabela_metricas_modelos_transfer_learning.csv")
print("- comparacao_f1_score_modelos.png")
print("- comparacao_acuracia_modelos.png")
print("- matriz_confusao_melhor_modelo.png")
print("- exemplos_erros_melhor_modelo.png")
print("- figura_exemplos_dataset.png")
print("- relatórios por classe em CSV")
print("- históricos de treinamento em CSV")
print("- curvas de treinamento em PNG")
print("- matrizes de confusão em PNG")

print("\nResultado final ordenado por F1-score macro:")
display(df_resultados_formatado)
```
