```python
# ============================================================
# Comparação entre MLP baseline e CNNs leves no Fashion-MNIST
# Modelos: MLP, CNN Leve
# Variações: dropout, otimizador e learning rate
# Google Colab
# ============================================================

import os
import gc
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

# ============================================================
# 1. Reprodutibilidade
# ============================================================

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

print("TensorFlow:", tf.__version__)

# ============================================================
# 2. Configurações gerais
# ============================================================

EPOCHS_MLP = 12
EPOCHS_CNN = 40
BATCH_SIZE = 128

# Se quiser acelerar testes, mude para True
USE_SUBSET = False
SUBSET_TRAIN = 15000
SUBSET_VAL = 3000
SUBSET_TEST = 3000

# ============================================================
# 3. Carregamento do dataset
# ============================================================

(x_train_full, y_train_full), (x_test, y_test) = fashion_mnist.load_data()

class_names = [
    "Camiseta/top", "Calça", "Pulôver", "Vestido", "Casaco",
    "Sandália", "Camisa", "Tênis", "Bolsa", "Bota"
]

print("Treino original:", x_train_full.shape)
print("Teste:", x_test.shape)

# ============================================================
# 4. Normalização e divisão treino/validação
# ============================================================

x_train_full = x_train_full.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

x_train = x_train_full[:50000]
y_train = y_train_full[:50000]

x_val = x_train_full[50000:]
y_val = y_train_full[50000:]

if USE_SUBSET:
    x_train = x_train[:SUBSET_TRAIN]
    y_train = y_train[:SUBSET_TRAIN]
    x_val = x_val[:SUBSET_VAL]
    y_val = y_val[:SUBSET_VAL]
    x_test = x_test[:SUBSET_TEST]
    y_test = y_test[:SUBSET_TEST]

print("Treino usado:", x_train.shape)
print("Validação usada:", x_val.shape)
print("Teste usado:", x_test.shape)

# Para MLP
x_train_mlp = x_train.reshape(-1, 28 * 28)
x_val_mlp = x_val.reshape(-1, 28 * 28)
x_test_mlp = x_test.reshape(-1, 28 * 28)

# Para modelos convolucionais
x_train_cnn = x_train.reshape(-1, 28, 28, 1)
x_val_cnn = x_val.reshape(-1, 28, 28, 1)
x_test_cnn = x_test.reshape(-1, 28, 28, 1)

# Labels one-hot
y_train_cat = to_categorical(y_train, 10)
y_val_cat = to_categorical(y_val, 10)
y_test_cat = to_categorical(y_test, 10)

# ============================================================
# 5. Visualização de exemplos
# ============================================================

plt.figure(figsize=(10, 5))
for i in range(12):
    plt.subplot(3, 4, i + 1)
    plt.imshow(x_train[i], cmap="gray")
    plt.title(class_names[y_train[i]])
    plt.axis("off")

plt.suptitle("Exemplos do dataset Fashion-MNIST")
plt.tight_layout()
plt.savefig("figura_exemplos_dataset.png", dpi=300)
plt.show()

# ============================================================
# 6. Funções auxiliares
# ============================================================

def criar_callbacks(patience=4, lr_patience=2):
    return [
        EarlyStopping(
            monitor="val_loss",
            patience=patience,
            restore_best_weights=True
        ),
        ReduceLROnPlateau(
            monitor="val_loss",
            factor=0.5,
            patience=lr_patience,
            min_lr=1e-6
        )
    ]


def obter_otimizador(nome, learning_rate):
    nome = nome.lower()
    if nome == "adam":
        return tf.keras.optimizers.Adam(learning_rate=learning_rate)
    if nome == "rmsprop":
        return tf.keras.optimizers.RMSprop(learning_rate=learning_rate)
    if nome == "sgd":
        return tf.keras.optimizers.SGD(
            learning_rate=learning_rate,
            momentum=0.9
        )
    raise ValueError(f"Otimizador não reconhecido: {nome}")


def calcular_metricas(nome, arquitetura, dropout, optimizer, learning_rate,
                      y_true, y_pred,
                      train_loss=None, train_acc=None,
                      val_loss=None, val_acc=None,
                      test_loss=None, test_acc=None):
    """Retorna dicionário com métricas de classificação e de conjunto (treino/val/teste)."""
    metricas = {
        "Modelo": nome,
        "Arquitetura": arquitetura,
        "Dropout": dropout,
        "Otimizador": optimizer,
        "Learning rate": learning_rate,
        "Acurácia Treino": train_acc,
        "Acurácia Validação": val_acc,
        "Acurácia Teste": accuracy_score(y_true, y_pred),
        "Perda Treino": train_loss,
        "Perda Validação": val_loss,
        "Perda Teste": test_loss,
        "Precisão macro": precision_score(y_true, y_pred, average="macro", zero_division=0),
        "Revocação macro": recall_score(y_true, y_pred, average="macro", zero_division=0),
        "F1-score macro": f1_score(y_true, y_pred, average="macro", zero_division=0)
    }
    return metricas


def plot_history(history, titulo, arquivo):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history["accuracy"], label="Treino")
    plt.plot(history.history["val_accuracy"], label="Validação")
    plt.title(f"Acurácia - {titulo}")
    plt.xlabel("Época")
    plt.ylabel("Acurácia")
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.subplot(1, 2, 2)
    plt.plot(history.history["loss"], label="Treino")
    plt.plot(history.history["val_loss"], label="Validação")
    plt.title(f"Perda - {titulo}")
    plt.xlabel("Época")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(arquivo, dpi=300)
    plt.show()


def plot_confusion(y_true, y_pred, titulo, arquivo):
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(9, 7))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=class_names,
        yticklabels=class_names
    )
    plt.title(titulo)
    plt.xlabel("Classe prevista")
    plt.ylabel("Classe real")
    plt.xticks(rotation=45, ha="right")
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.savefig(arquivo, dpi=300)
    plt.show()


def nome_arquivo_seguro(texto):
    return (
        texto.lower()
        .replace(" ", "_")
        .replace("+", "mais")
        .replace("/", "_")
        .replace("-", "_")
        .replace(".", "_")
    )


def avaliar_conjuntos(modelo, x_train, y_train, x_val, y_val, x_test, y_test):
    """Avalia o modelo nos conjuntos de treino, validação e teste.
       Retorna as perdas e acurácias."""
    train_loss, train_acc = modelo.evaluate(x_train, y_train, verbose=0)
    val_loss, val_acc = modelo.evaluate(x_val, y_val, verbose=0)
    test_loss, test_acc = modelo.evaluate(x_test, y_test, verbose=0)
    return train_loss, train_acc, val_loss, val_acc, test_loss, test_acc


# ============================================================
# 7. Modelo baseline - MLP
# ============================================================

mlp_model = models.Sequential([
    layers.Input(shape=(784,)),
    layers.Dense(256, activation="relu", kernel_regularizer=l2(1e-4)),
    layers.Dropout(0.4),
    layers.Dense(128, activation="relu", kernel_regularizer=l2(1e-4)),
    layers.Dropout(0.4),
    layers.Dense(10, activation="softmax")
])

mlp_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

print("\nResumo do modelo MLP baseline:")
mlp_model.summary()

print("\nTreinando MLP baseline...")

history_mlp = mlp_model.fit(
    x_train_mlp, y_train_cat,
    validation_data=(x_val_mlp, y_val_cat),
    epochs=EPOCHS_MLP,
    batch_size=BATCH_SIZE,
    callbacks=criar_callbacks(),
    verbose=1
)

# Avalia explicitamente treino, validação e teste
train_loss, train_acc, val_loss, val_acc, test_loss, test_acc = avaliar_conjuntos(
    mlp_model, x_train_mlp, y_train_cat, x_val_mlp, y_val_cat, x_test_mlp, y_test_cat
)

mlp_probs = mlp_model.predict(x_test_mlp)
mlp_pred = np.argmax(mlp_probs, axis=1)

resultados = []
predicoes = {}
historicos = {}

resultados.append(
    calcular_metricas(
        nome="MLP baseline",
        arquitetura="MLP",
        dropout=0.4,
        optimizer="Adam",
        learning_rate=0.001,
        y_true=y_test,
        y_pred=mlp_pred,
        train_loss=train_loss, train_acc=train_acc,
        val_loss=val_loss, val_acc=val_acc,
        test_loss=test_loss, test_acc=test_acc
    )
)

predicoes["MLP baseline"] = mlp_pred
historicos["MLP baseline"] = history_mlp

print("\nRelatório MLP baseline (teste):")
print(classification_report(y_test, mlp_pred, target_names=class_names, zero_division=0))

pd.DataFrame(
    classification_report(y_test, mlp_pred, target_names=class_names,
                          output_dict=True, zero_division=0)
).transpose().to_csv("relatorio_por_classe_mlp_baseline.csv")

plot_history(history_mlp, "MLP baseline", "curvas_mlp_baseline.png")
plot_confusion(y_test, mlp_pred, "Matriz de Confusão - MLP baseline",
               "matriz_confusao_mlp_baseline.png")

# ============================================================
# 8. Data augmentation (leve, para a CNN leve)
# ============================================================

data_augmentation_cnn = tf.keras.Sequential([
    layers.RandomZoom(0.05),
    layers.RandomFlip("horizontal")
], name="data_augmentation")

# ============================================================
# 9. Arquitetura CNN leve do zero
# ============================================================

def criar_cnn_leve(dropout_rate, optimizer_name, learning_rate):
    model = models.Sequential([
        layers.Input(shape=(28, 28, 1)),

        # Bloco 1
        layers.Conv2D(32, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.Conv2D(32, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(dropout_rate),

        # Bloco 2
        layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(dropout_rate),

        # Bloco 3
        layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(dropout_rate),

        # Cabeça
        layers.GlobalAveragePooling2D(),
        layers.Dense(128, activation='relu', kernel_regularizer=l2(1e-4)),
        layers.BatchNormalization(),
        layers.Dropout(dropout_rate * 2),
        layers.Dense(10, activation='softmax')
    ])
    model.compile(
        optimizer=obter_otimizador(optimizer_name, learning_rate),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# ============================================================
# 10. Experimentos com CNN leve
# ============================================================

experimentos_cnn = [
    {"nome": "CNN Leve - Drop 0.20 - Adam - LR 1e-3",
     "dropout": 0.20, "optimizer": "Adam", "learning_rate": 1e-3},
    {"nome": "CNN Leve - Drop 0.30 - Adam - LR 5e-4",
     "dropout": 0.30, "optimizer": "Adam", "learning_rate": 5e-4},
    {"nome": "CNN Leve - Drop 0.20 - RMSprop - LR 1e-3",
     "dropout": 0.20, "optimizer": "RMSprop", "learning_rate": 1e-3},
    {"nome": "CNN Leve - Drop 0.30 - RMSprop - LR 5e-4",
     "dropout": 0.30, "optimizer": "RMSprop", "learning_rate": 5e-4}
]

for exp in experimentos_cnn:
    print("\n" + "=" * 80)
    print("Treinando:", exp["nome"])
    print("=" * 80)

    tf.keras.backend.clear_session()
    gc.collect()

    # Cria a CNN pura
    cnn = criar_cnn_leve(
        dropout_rate=exp["dropout"],
        optimizer_name=exp["optimizer"],
        learning_rate=exp["learning_rate"]
    )

    # Encapsula com data augmentation
    inputs = layers.Input(shape=(28, 28, 1))
    x = data_augmentation_cnn(inputs)
    outputs = cnn(x)
    full_model = models.Model(inputs, outputs)
    full_model.compile(
        optimizer=obter_otimizador(exp["optimizer"], exp["learning_rate"]),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    print("\nResumo do modelo:")
    full_model.summary()

    cb = criar_callbacks(patience=8, lr_patience=4)

    history = full_model.fit(
        x_train_cnn, y_train_cat,
        validation_data=(x_val_cnn, y_val_cat),
        epochs=EPOCHS_CNN,
        batch_size=BATCH_SIZE,
        callbacks=cb,
        verbose=1
    )

    # Avaliação explícita nos três conjuntos
    train_loss, train_acc, val_loss, val_acc, test_loss, test_acc = avaliar_conjuntos(
        full_model, x_train_cnn, y_train_cat, x_val_cnn, y_val_cat, x_test_cnn, y_test_cat
    )

    probs = full_model.predict(x_test_cnn)
    y_pred = np.argmax(probs, axis=1)

    resultados.append(
        calcular_metricas(
            nome=exp["nome"],
            arquitetura="CNN Leve",
            dropout=exp["dropout"],
            optimizer=exp["optimizer"],
            learning_rate=exp["learning_rate"],
            y_true=y_test,
            y_pred=y_pred,
            train_loss=train_loss, train_acc=train_acc,
            val_loss=val_loss, val_acc=val_acc,
            test_loss=test_loss, test_acc=test_acc
        )
    )

    predicoes[exp["nome"]] = y_pred
    historicos[exp["nome"]] = history

    print(f"\nRelatório por classe - {exp['nome']} (teste):")
    print(classification_report(y_test, y_pred, target_names=class_names, zero_division=0))

    nome_seguro = nome_arquivo_seguro(exp["nome"])
    pd.DataFrame(
        classification_report(y_test, y_pred, target_names=class_names,
                              output_dict=True, zero_division=0)
    ).transpose().to_csv(f"relatorio_por_classe_{nome_seguro}.csv")

    plot_history(history, exp["nome"], f"curvas_{nome_seguro}.png")
    plot_confusion(y_test, y_pred, f"Matriz de Confusão - {exp['nome']}",
                   f"matriz_confusao_{nome_seguro}.png")

    del cnn, full_model
    gc.collect()

# ============================================================
# 11. Tabela final de resultados (explicitando treino/val/teste)
# ============================================================

df_resultados = pd.DataFrame(resultados)
df_resultados = df_resultados.sort_values(by="F1-score macro", ascending=False).reset_index(drop=True)

# Formatação para exibição
colunas_numericas = ["Acurácia Treino", "Acurácia Validação", "Acurácia Teste",
                     "Perda Treino", "Perda Validação", "Perda Teste",
                     "Precisão macro", "Revocação macro", "F1-score macro"]
df_formatado = df_resultados.copy()
for col in colunas_numericas:
    df_formatado[col] = df_formatado[col].map(lambda x: f"{x:.4f}" if pd.notnull(x) else "N/A")
# Dropout e LR também podem ser formatados
df_formatado["Dropout"] = df_formatado["Dropout"].map(lambda x: f"{x:.4f}")
df_formatado["Learning rate"] = df_formatado["Learning rate"].map(lambda x: f"{x:.4f}")

print("\n=== TABELA COMPARATIVA FINAL ===")
display(df_formatado[["Modelo", "Arquitetura", "Acurácia Treino", "Acurácia Validação",
                      "Acurácia Teste", "F1-score macro"]])
df_resultados.to_csv("tabela_final_comparativa.csv", index=False)

# ============================================================
# 12. Gráficos comparativos
# ============================================================

plt.figure(figsize=(14, 6))
df_plot = df_resultados.copy()
df_plot["Modelo curto"] = df_plot["Modelo"].str.replace(" - ", "\n", regex=False)
sns.barplot(data=df_plot, x="Modelo curto", y="F1-score macro", hue="Arquitetura")
plt.title("Comparação dos modelos por F1-score macro")
plt.xlabel("Modelo")
plt.ylabel("F1-score macro")
plt.ylim(0, 1)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Arquitetura")
plt.tight_layout()
plt.savefig("comparacao_f1_score_modelos.png", dpi=300)
plt.show()

plt.figure(figsize=(14, 6))
sns.barplot(data=df_plot, x="Modelo curto", y="Acurácia Teste", hue="Arquitetura")
plt.title("Comparação dos modelos por acurácia (teste)")
plt.xlabel("Modelo")
plt.ylabel("Acurácia")
plt.ylim(0, 1)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Arquitetura")
plt.tight_layout()
plt.savefig("comparacao_acuracia_modelos.png", dpi=300)
plt.show()

# ============================================================
# 13. Melhor modelo e análise de erros
# ============================================================

melhor_modelo = df_resultados.iloc[0]["Modelo"]
melhor_pred = predicoes[melhor_modelo]

print("\nMelhor modelo encontrado:")
print(melhor_modelo)

plot_confusion(y_test, melhor_pred,
               f"Matriz de Confusão - Melhor modelo: {melhor_modelo}",
               "matriz_confusao_melhor_modelo.png")

erros = np.where(melhor_pred != y_test)[0]
plt.figure(figsize=(12, 8))
for i, idx in enumerate(erros[:12]):
    plt.subplot(3, 4, i + 1)
    plt.imshow(x_test[idx], cmap="gray")
    plt.title(f"Real: {class_names[y_test[idx]]}\nPred: {class_names[melhor_pred[idx]]}", fontsize=9)
    plt.axis("off")
plt.suptitle(f"Exemplos de erros - {melhor_modelo}")
plt.tight_layout()
plt.savefig("exemplos_erros_melhor_modelo.png", dpi=300)
plt.show()

# ============================================================
# 14. Salvamento dos históricos
# ============================================================

for nome_modelo, history in historicos.items():
    nome_seguro = nome_arquivo_seguro(nome_modelo)
    pd.DataFrame(history.history).to_csv(f"historico_treinamento_{nome_seguro}.csv", index=False)

# ============================================================
# 15. Resumo final
# ============================================================

print("\nArquivos principais gerados:")
print("- tabela_final_comparativa.csv")
print("- comparacao_f1_score_modelos.png")
print("- comparacao_acuracia_modelos.png")
print("- matriz_confusao_melhor_modelo.png")
print("- exemplos_erros_melhor_modelo.png")
print("- figura_exemplos_dataset.png")
print("- relatórios por classe em CSV")
print("- históricos de treinamento em CSV")
print("- curvas de treinamento em PNG")
print("- matrizes de confusão em PNG")

print("\nResultado final ordenado por F1-score macro (teste):")
display(df_formatado[["Modelo", "Arquitetura", "Acurácia Treino", "Acurácia Validação",
                      "Acurácia Teste", "F1-score macro"]])
```
